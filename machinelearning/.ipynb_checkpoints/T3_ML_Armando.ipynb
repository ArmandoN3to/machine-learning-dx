```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns   

```

```python
path='C:/Users/arman/codigos_github/machine-learning-dx/machinelearning/datasets/'
candidatura_2022=pd.read_csv(path+'consulta_cand_2022_BRASIL.csv',encoding='latin-1',sep=';')#consulta_cand_2022_BRASILsulta_cand_2022_BRASIL.csv
```

```python
candidatura_2022.head()
```

```python
candidatura_2022.info()
```

```python
labelenconder_tipo = LabelEncoder()
labelenconder_bairro = LabelEncoder()

candidatura_2022['SG_PARTIDO'] = labelenconder_tipo.fit_transform(candidatura_2022['SG_PARTIDO'])
candidatura_2022['NM_PARTIDO'] = labelenconder_bairro.fit_transform(candidatura_2022['NM_PARTIDO']) 
```

```python
X_candidatura=candidatura_2022[['NR_PARTIDO','SG_PARTIDO']]
y_candidatura=candidatura_2022['NM_PARTIDO']
```

```python
X_train_candidatura, X_test_candidatura, y_train_candidatura, y_test_candidatura = train_test_split(X_candidatura, y_candidatura,
                                                                                 test_size=0.4, shuffle=True,random_state=0)
X_train_candidatura.shape,X_test_candidatura.shape,y_train_candidatura.shape,y_test_candidatura.shape
```

```python
X_candidatura
```

```python
modelo_tree= DecisionTreeClassifier(random_state=0)
modelo_tree.fit(X_train_candidatura,y_train_candidatura)
```

```python
y_pred_candidatura=modelo_tree.predict(X_test_candidatura)
acuracia_tree= accuracy_score(y_test_candidatura,y_pred_candidatura)
print(f"Acuracia da arvore de decisao: {acuracia_tree}")
```

```python
print("Relatorio de Classificaçao: ")
print(classification_report(y_test_candidatura,y_pred_candidatura))
matriz_conf_knn=confusion_matrix(y_test_candidatura,y_pred_candidatura)
sns.heatmap(matriz_conf_knn, annot=True, fmt='d', cmap="Blues")
plt.xlabel(" Matriz confusao")
```
