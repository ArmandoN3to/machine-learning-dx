# CLASSIFICAÇÃO

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns   

```

```python
path='C:/Users/arman/codigos_github/machine-learning-dx/machinelearning/datasets/'
candidatura_2022=pd.read_csv(path+'consulta_cand_2022_BRASIL.csv',encoding='latin-1',sep=';')#consulta_cand_2022_BRASILsulta_cand_2022_BRASIL.csv
```

```python
candidatura_2022.head()
```

```python
candidatura_2022.info()
```

```python
labelenconder_sg = LabelEncoder()
labelenconder_nm = LabelEncoder()

candidatura_2022['SG_PARTIDO'] = labelenconder_sg.fit_transform(candidatura_2022['SG_PARTIDO'])
candidatura_2022['NM_PARTIDO'] = labelenconder_nm.fit_transform(candidatura_2022['NM_PARTIDO']) 
```

```python
X_candidatura=candidatura_2022[['NR_PARTIDO','SG_PARTIDO']]
y_candidatura=candidatura_2022['NM_PARTIDO']
```

```python
X_train_candidatura, X_test_candidatura, y_train_candidatura, y_test_candidatura = train_test_split(X_candidatura, y_candidatura,
                                                                                 test_size=0.4, shuffle=True,random_state=0)
X_train_candidatura.shape,X_test_candidatura.shape,y_train_candidatura.shape,y_test_candidatura.shape
```

```python
X_candidatura
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
```

```python
modelos_classifier = {
     'DecisionTreeClassifier': DecisionTreeClassifier(random_state = 0),
     'RandomForestClassifier' : RandomForestClassifier(n_estimators = 200,random_state = 0),
     'LogisticRegression': LogisticRegression(random_state = 0),
     'SVC' : SVC(random_state = 0),
     'KNN' : KNN(n_neighbors = 6)
}
```

```python
resulatdos_classifier = {
    'Modelo': [],
    'Acuracia': []
}
```

```python
for nome_modelo, modelo in  modelos_classifier.items():
    modelo.fit(X_train_candidatura,y_train_candidatura)
    y_preds_classifier = modelo.predict(X_test_candidatura)
    accuracy_classifier = accuracy_score(y_test_candidatura,y_preds_classifier)
    resulatdos_classifier['Modelo'].append(nome_modelo)
    resulatdos_classifier['Acuracia'].append(accuracy_classifier)
    print(f'\n==== {nome_modelo} =====')
    print(f'Acurácia: {accuracy_classifier:.4f}')
    print('Relatório de classificação')
    print(classification_report(y_test_candidatura,y_preds_classifier, zero_division=0))
    print('Matriz de confusão')
    print(confusion_matrix(y_test_candidatura,y_preds_classifier))
    print('=' * 80)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data = resulatdos_classifier,
    x = 'Modelo',
    y = 'Acuracia',
    hue = 'Modelo',
    palette = 'viridis',
    legend = False
)

plt.title('Comparação de modelos ')
for i, v in enumerate(resulatdos_classifier['Acuracia']):
    plt.text(i,v + 0.005, f'{v:.3f}', ha = 'center', fontweight = 'bold')
plt.xticks(rotation = 40, ha = 'right')
##plt.subplots.adjust(bottom = 0.25)
plt.show()
```

# REGREÇÃO

```python
df_estoque_DPF = pd.read_csv(path + 'EstoqueDPF.csv',sep=';')
df_estoque_DPF.info()
```

```python
df_estoque_DPF.head()
```

#### Selecionar valores numericos

```python
colunas_numericas = ['Valor do Estoque', 'Quantidade do Estoque']
for coluna in colunas_numericas:
    df_estoque_DPF[coluna] = pd.to_numeric(df_estoque_DPF[coluna].str.replace(',', '.'), errors='coerce')
```

```python
df_numerico = df_estoque_DPF.select_dtypes(include=['float64', 'int64'])
```

#### Correlaçao

```python
correlacao = df_numerico.corr()
print(correlacao)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
modelos_de_regressao = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(n_estimators=100, random_state = 0),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state = 0)
}
```

```python
resultados_rg = {
    'Modelo': [],
    'Score': []
}
```

```python
X_estoque = df_estoque_DPF.drop(['Titulo/Contrato','Vencimento do Titulo/Contrato','Mes do Estoque','Classe da Carteira','Tipo de Divida'],axis =1 )
y_estoque = df_estoque_DPF['Quantidade do Estoque']
X_estoque.shape, y_estoque.shape
```

```python
X_train_estoque,X_test_estoque, y_train_estoque,y_test_estoque = train_test_split(X_estoque,y_estoque,
    test_size = 0.30,
    random_state = 0
)
```

```python
for nome_modelo, modelo in  modelos_de_regressao.items():
    modelo.fit(X_train_estoque,y_train_estoque)
    y_preds_regression = modelo.predict(X_test_estoque)
    mse_regression = mean_squared_error(y_test_estoque,y_preds_regression)
    r2_score_regression = r2_score(y_test_estoque,y_preds_regression)
    resultados_rg['Modelo'].append(nome_modelo)
    resultados_rg['Score'].append(r2_score_regression)

    print(f'\n==== {nome_modelo} =====')
    print('MSE: ', mse_regression)
    print('r2_score: ', r2_score_regression)
    print('-' * 80)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data = resultados_rg,
    x = 'Modelo',
    y = 'Score',
    hue = 'Modelo',
    palette = 'viridis',
    legend = False
)

# plt.title('Comparação de R2 SCORE - ESTOQUE')
for i, v in enumerate(resultados_rg['Score']):
    plt.text(i,v + 0.005, f'{v:.3f}', ha = 'center', fontweight = 'bold')
plt.xticks(rotation = 40, ha = 'right')
plt.ylim(0.90,1.0)
##plt.subplots.adjust(bottom = 0.25)
plt.show()
```
