```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
digitos = load_digits()
len(digitos.feature_names)
# digitos.target
# digitos.target_names
#digitos.images
```

```python
fig=plt.figure(figsize=(8, 6))
fig.subplots_adjust(left=0, right=1, bottom=0,top=1,
                    hspace=0.05,wspace=0.05)
for i in range(25):
    ax=fig.add_subplot(5,5,i+1,xticks=[],yticks=[])
    ax.imshow(digitos.images[i],cmap=plt.cm.gray,interpolation = 'lanczos')
```

```python
X_dig = digitos.data
y_dig = digitos.target # Dividindo o dataset em treino e teste
```

```python
X_train_dig, X_test_dig,y_train_dig, y_test_dig=train_test_split(X_dig, y_dig, test_size=0.3, random_state=0)
```

```python
modelo_knn=KNN(n_neighbors=5)
modelo_knn.fit(X_train_dig,y_train_dig)
```

```python
pred_dig=modelo_knn.predict(X_test_dig)
acuracia_knn=modelo_knn.score(X_test_dig,y_test_dig)
print('Acuracia do KNN:',acuracia_knn)
```

```python
matriz_conf_knn=metrics.confusion_matrix(y_test_dig,pred_dig)
sns.heatmap(matriz_conf_knn, annot=True, fmt='d', cmap="Blues")
plt.xlabel(" Matriz confusao")
```

```python
relat_class_knn=metrics.classification_report(y_test_dig,pred_dig)
print("Relatorio de classificaçao")
print(relat_class_knn)
```

## Navy 

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
resultados = {
    'Modelo': [],
    'Acuracia': []
}
```

```python
modelos_nb = {
    "GaussianNB": GaussianNB(),
    "BernoulliNB":BernoulliNB(),
    "MultinomialNB": MultinomialNB()
}
```

```python
for nome_modelo, modelo in  modelos_nb.items():
    modelo.fit(X_train_dig, y_train_dig)
    y_preds_dig = modelo.predict(X_test_dig)
    acc = accuracy_score(y_test_dig, y_preds_dig)
    resultados['Modelo'].append(nome_modelo)
    resultados['Acuracia'].append(acc)


    print(f'\n==== {nome_modelo} =====')
    print(f'Acurácia: {acc:.4f}')
    print('Relatório de classificação')
    print(classification_report(y_test_dig, y_preds_dig, zero_division=0))
    print('Matriz de confusão')
    print(confusion_matrix(y_test_dig,y_preds_dig))
    print('-' * 80)
```

## SVM

```python
modelos_svm = {
    "SVM (Linear)": SVC(kernel='linear',C=1.0,random_state=0),
    "SVM (RBF)": SVC(kernel='rbf', C=1.0, random_state=0), # melhor
    "SVM (Poly)": SVC(kernel='poly', C=1.0, random_state=0)
}
```

```python
for nome_modelo, modelo in  modelos_svm.items():
    modelo.fit(X_train_dig, y_train_dig)
    y_preds_dig = modelo.predict(X_test_dig)
    acc = accuracy_score(y_test_dig, y_preds_dig)
    resultados['Modelo'].append(nome_modelo)
    resultados['Acuracia'].append(acc)


    print(f'\n==== {nome_modelo} =====')
    print(f'Acurácia: {acc:.4f}')
    print('Relatório de classificação')
    print(classification_report(y_test_dig, y_preds_dig, zero_division=0))
    print('Matriz de confusão')
    print(confusion_matrix(y_test_dig,y_preds_dig))
    print('-' * 80)
```

## Random Forest

```python
modelos_rf={
    'RF': RandomForestClassifier(n_estimators=100,random_state=0)
}
```

```python
for nome_modelo, modelo in  modelos_rf.items():
    modelo.fit(X_train_dig, y_train_dig)
    y_preds_dig = modelo.predict(X_test_dig)
    acc = accuracy_score(y_test_dig, y_preds_dig)
    resultados['Modelo'].append(nome_modelo)
    resultados['Acuracia'].append(acc)


    print(f'\n==== {nome_modelo} =====')
    print(f'Acurácia: {acc:.4f}')
    print('Relatório de classificação')
    print(classification_report(y_test_dig, y_preds_dig, zero_division=0))
    print('Matriz de confusão')
    print(confusion_matrix(y_test_dig,y_preds_dig))
    print('-' * 80)
```

```python
df_resultados =  pd.DataFrame(resultados)
```

```python
print('\n ============ Resumo dos resultados =========')
print(df_resultados)    
```
