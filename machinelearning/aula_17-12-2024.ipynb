## Funçao para 3 imagens

```python
def plot_3img(im1,tit1,im2,tit2,im3,tit3,tam):
    im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize=tam)
    plt.subplot(1,3,1),plt.imshow(im1),plt.title(tit1)
    plt.xticks([]),plt.yticks([])
    
    im2=cv2.cvtColor(im2,cv2.COLOR_BGR2RGB)
    plt.subplot(1,3,2),plt.imshow(im2),plt.title(tit2)
    plt.xticks([]),plt.yticks([])
    
    im3=cv2.cvtColor(im3,cv2.COLOR_BGR2RGB)
    plt.subplot(1,3,3),plt.imshow(im3),plt.title(tit3)
    plt.xticks([]),plt.yticks([])
    plt.show
```

```python
k1=np.array([[-1,-1,-1],
             [-1,9,-1],
             [-1,-1,-1]])
k2=np.array([[1,0,0],
             [0,0,0],
             [0,0,-1]])

k3=np.array([[0,0,0],
             [0,1,0],
             [0,0,0]])

k4=np.array([[0,-1,0],
             [-1,5,-1],
             [0,-1,0]])

k5=np.array([[0,1,0],
             [1,-4,1],
             [0,1,0]])

k6=np.array([[1,0,-1],
             [0,0,0],
             [-1,0,1]])

k7=np.array([[-1,0,1],
             [-2,0,2],
             [-1,0,1]])
```

```python
#passar imagem e nome , kernel
def aplicar_filtro(im1,tit1,kernel,tam):
    
    im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=tam)
    plt.subplot(1,2,1),plt.imshow(im1),plt.title('Original')
    plt.xticks([]),plt.yticks([])
    plt.show()
    
    im1=cv2.filter2D(im1,-1,kernel)
    plt.figure(figsize=tam)
    plt.subplot(1,2,2),plt.imshow(im1),plt.title(tit1)
    plt.xticks([]),plt.yticks([])
    plt.show()
```
